```python
# initiate Trino client

from trino.dbapi import connect


trino_connection = connect(
    host="trino",
    port=8080,
    user="iceberg",
)
trino = trino_connection.cursor()
```

```python
# select all data in the table

rows = trino.execute('SELECT * FROM iceberg.default.bank_transfers').fetchall()

for row in rows:
    print(row)
```

```python
# same as the above, but with explicit mention of the data table

rows = trino.execute('SELECT * FROM iceberg.default."bank_transfers$data"').fetchall()

for row in rows:
    print(row)
```

```python
# query table history

cols = trino.execute('SHOW COLUMNS FROM iceberg.default."bank_transfers$history"').fetchall()
rows = trino.execute('SELECT * FROM iceberg.default."bank_transfers$history"').fetchall()

for col in cols:
    print(col)

print("")
    
for row in rows:
    print(row)
```

```python
# query table snapshots

cols = trino.execute('SHOW COLUMNS FROM iceberg.default."bank_transfers$snapshots"').fetchall()
rows = trino.execute("""
SELECT * FROM iceberg.default."bank_transfers$snapshots"
ORDER BY committed_at DESC
""").fetchall()


for col in cols:
    print(col)

print("")
    
for row in rows:
    print(row)
```

```python
# query a particular snapshot

rows = trino.execute("""
SELECT * FROM iceberg.default.bank_transfers
FOR VERSION AS OF 1091088526575766821
""").fetchall()

for row in rows:
    print(row)
```
